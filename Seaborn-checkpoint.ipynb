```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Line plot

```python
fmri= sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event', marker=True)
plt.show()
```

# Bar plot 

```python
sns.barplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

```python
sns.barplot(x='timepoint', y='signal', data=fmri, hue='event')
plt.show()
```

```python
sns.barplot(x='timepoint', y='signal', data=fmri, palette='Blues_d')
plt.show()
```

```python
sns.barplot(x='timepoint', y='signal', data=fmri, palette='rocket')
plt.show()
```

```python
sns.barplot(x='timepoint', y='signal', data=fmri, palette='vlag')
plt.show()
```

```python
sns.barplot(x='timepoint', y='signal', data=fmri, color='r')
plt.show()
```

# Scatter plot 

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri, hue='event', style='event')
plt.show()
```

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri, hue='region', style='event')
plt.show()
```

# Histogram

```python
sns.histplot(fmri['timepoint'])
plt.show()
```

```python
sns.distplot(fmri['timepoint'], hist=False)
plt.show()
```

```python
sns.distplot(fmri['timepoint'], color='r')
plt.show()
```

```python
sns.distplot(fmri['timepoint'], color='r', kde=False, bins=5)
plt.show()
```

```python
sns.distplot(fmri['timepoint'], color='r', kde=False, bins=5, vertical=True)
plt.show()
```

# Joint plot

```python
sns.jointplot(x='timepoint', y='signal', data=fmri, color='y')
plt.show()
```

```python
sns.jointplot(x='timepoint', y='signal', data=fmri, color='y',kind='reg')
plt.show()
```

# Box plot

```python
sns.boxplot=(x='timepoint', y='event', data=fmri)
plt.show()
```
